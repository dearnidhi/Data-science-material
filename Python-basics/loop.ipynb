1. For Loops:
A for loop in Python is used to iterate over a sequence (such as a list, tuple, string, or range) or any iterable object.

Syntax:

```python
for item in iterable:
    # Code block to execute
```

Example 1: Iterating over a list

```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

Output:

```python
apple
banana
cherry
```

Example 2: Iterating over a string

```python
for char in "hello":
    print(char)
```

Output:

```python
h
e
l
l
o
```

Example 3: Using range function

```python
for num in range(1, 10):  # range from 1 to 4 (5 is exclusive)
    print(num)
```

Output:

```python
1
2
3
4
```

Example 4: Iterating over a tuple of tuples

```python
coordinates = [(1, 2), (3, 4), (5, 6)]
for x, y in coordinates:
    print(f"X: {x}, Y: {y}")
```

Output:

```python
X: 1, Y: 2
X: 3, Y: 4
X: 5, Y: 6
```

2. While Loops:
A while loop in Python repeats a block of code as long as a specified condition is true.

Syntax:

```python
while condition:
    # Code block to execute
    # This continues until condition becomes False
```

Example 1: Simple while loop

**Step-by-Step Execution:**
First Iteration:

count is 0 (initial value).<br/>
The condition count < 5 is true (0 < 5).<br/>
print(count) prints 0.<br/>
count += 1 updates count to 1.<br/>

Second Iteration:

count is 1.<br/>
The condition count < 5 is true (1 < 5).<br/>
print(count) prints 1.<br/>
count += 1 updates count to 2.<br/>

Third Iteration:

count is 2.<br/>
The condition count < 5 is true (2 < 5).<br/>
print(count) prints 2.<br/>
count += 1 updates count to 3.<br/>

Fourth Iteration:

count is 3.<br/>
The condition count < 5 is true (3 < 5).<br/>
print(count) prints 3.<br/>
count += 1 updates count to 4.<br/>

Fifth Iteration:

count is 4.<br/>
The condition count < 5 is true (4 < 5).<br/>
print(count) prints 4.<br/>
count += 1 updates count to 5.<br/>

Loop Termination:

count is 5.<br/>
The condition count < 5 is false (5 < 5 is not true).<br/>
The loop terminates and exits.<br/>

```python
count = 0
while count < 5:
    print(count)
    count += 1
```

Output:

```python
0
1
2
3
4
```

Example 3: Using 'continue' to skip an iteration

```python
count = 0
while count < 5:
    count += 1
    if count == 3:
        continue
    print(count)
```

```python
1
2
4
5
```

3. Loop Control Statements:

Python provides several loop control statements to modify the execution of loops:

a. break terminates the loop entirely.

b. continue skips the rest of the code inside the loop for the current iteration and continues with the next iteration.

c. pass is a null operation; it does nothing when executed. It can be used as a placeholder.

Example using break:

```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    if fruit == "banana":
        break
    print(fruit)
```

Example using continue:

Step-by-Step Execution:

First Iteration (num = 1):

The condition num % 2 == 0 evaluates to False (1 is not even).
print(num) prints 1.

Second Iteration (num = 2):

The condition num % 2 == 0 evaluates to True (2 is even).
The continue statement is executed, skipping the print(num) statement.
The loop moves to the next iteration without printing anything.


Third Iteration (num = 3):

The condition num % 2 == 0 evaluates to False (3 is not even).
print(num) prints 3.


Fourth Iteration (num = 4):

The condition num % 2 == 0 evaluates to True (4 is even).
The continue statement is executed, skipping the print(num) statement.
The loop moves to the next iteration without printing anything.


Fifth Iteration (num = 5):

The condition num % 2 == 0 evaluates to False (5 is not even).
print(num) prints 5.


Summary:
The for loop iterates over the numbers from 1 to 5.
For each number, the loop checks if the number is even.
If the number is even, the continue statement skips the rest of the code inside the loop for that iteration.
If the number is odd, it prints the number.
As a result, the code prints the odd numbers: 1, 3, and 5.

```python
for num in range(1, 6):
    if num % 2 == 0:
        continue
    print(num)
```

Example using pass:

```python
for letter in 'hello':
    if letter == 'l':
        pass  # do nothing for 'l'
    else:
        print('Skip Letter :', letter)
```

4. Nested Loops:
Python allows nesting of loops (i.e., putting loops inside another loop). This can be useful in various scenarios, such as iterating through matrices or performing complex data manipulations.

Example: Nested for loop

```python
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i} * {j} = {i * j}")
```

5. Loop Else Clause:
Python allows an optional else clause with loops which is executed when the loop completes normally (i.e., without encountering a break statement).

Example: for loop with else clause



```python
fruits = ["apple", "banana", "cherry","orange"]
for fruit in fruits:
    if fruit == "orange":
        print("Found orange!")
        break
else:
    print("No oranges found.")
```

Example: while loop with else clause

Step-by-Step Execution:
First Iteration:

count is 0.
The condition count < 3 is true (0 < 3).
print(count) prints 0.
count += 1 updates count to 1.
Second Iteration:

count is 1.
The condition count < 3 is true (1 < 3).
print(count) prints 1.
count += 1 updates count to 2.
Third Iteration:

count is 2.
The condition count < 3 is true (2 < 3).
print(count) prints 2.
count += 1 updates count to 3.


Loop Termination:

count is 3.
The condition count < 3 is false (3 < 3 is not true).
The loop exits.


Else Block Execution:

After the loop exits, the else block executes.
print("Loop completed successfully") prints the message "Loop completed successfully".


Summary:
The while loop runs three times, printing the values 0, 1, and 2.
When count reaches 3, the loop condition count < 3 becomes false, and the loop exits.
The else block is then executed, printing the message "Loop completed successfully".


```python
count = 0
while count < 3:
    print(count)
    count += 1
else:
    print("Loop completed successfully")
```
