```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.keys()
```

```python
print(iris['DESCR'])
```

```python
df = pd.DataFrame(iris['data'], columns = iris['feature_names'])
df.head()
```

```python
df['target'] = iris['target']
```

```python
df.head()
```

```python
df.sample(10)
```

```python
X = iris['data']
y = iris['target']
```

```python
y
```

```python
X
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_cat = to_categorical(y)
```

```python
y_cat
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size = 0.33)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(patience = 10)
```

```python
model = Sequential()

model.add(Dense(4, activation = 'relu'))
model.add(Dense(4, activation = 'relu'))
model.add(Dense(4, activation = 'relu'))
model.add(Dense(3, activation = 'softmax'))

model.compile(loss = 'categorical_crossentropy', optimizer='adam', metrics = ['accuracy'])
```

```python
model.fit(X_train, y_train, epochs = 500, validation_data = (X_test, y_test), callbacks = [early_stop])
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df
```

```python
loss_df[['val_loss', 'loss']].plot()
```

```python
loss_df[['val_accuracy', 'accuracy']].plot()
```

```python
predictions = model.predict(X_test)
```

```python
predictions = np.argmax(predictions, axis = 1)
```

```python
actual_y = np.argmax(y_test, axis = 1)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(actual_y, predictions))
```
