In **Decorators**, functions are passed as an argument into another function and then called inside the wrapper function.


#Here's what happens when you call say_hello():

The wrapper function is executed.

It prints "Something is happening before the function is called."

It calls the say_hello function using func().

The say_hello function executes and prints "Hello!"

The wrapper function continues execution and prints "Something is happening after the function is called."


```python
#Basic Concept of Decorators
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

def say_hello():
    print("Hello!")

say_hello = my_decorator(say_hello)

say_hello()

```

#Using the @ Syntax
Python provides a cleaner way to apply decorators using the @ syntax. Here's the same example using this syntax:

```python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

#Decorators with Arguments
If you want your decorator to accept arguments, you need to add another layer of functions. Here's an example:

Here's a summary of which function is called first and the print statements:

The repeat function is called first with num_times=3.

The repeat function returns decorator_repeat.

The decorator_repeat function is called with say_hello.

The decorator_repeat function returns wrapper.

The say_hello() function (which is now wrapper) is called.

The wrapper function calls say_hello 3 times, each time printing "Hello!".

```python
def repeat(num_times):
    def decorator_repeat(func):
        def wrapper(*args, **kwargs):
            for _ in range(num_times):
                func(*args, **kwargs)
        return wrapper
    return decorator_repeat

@repeat(num_times=3)
def say_hello():
    print("Hello!")

say_hello()
```

#Step-by-Step Execution


* say_hello("Alice"): This invokes the wrapper function because say_hello has been replaced by wrapper due to the decorator.

* Inside wrapper:

Prints "Something is happening before the function is called."


Calls the original say_hello function with name="Alice"

result = func(*args, **kwargs)

This translates to:

result = say_hello("Alice")

Which executes:

print(f"Hello, {name}!")

Resulting in the output:

Hello, Alice!

Prints "Something is happening after the function is called."

Returns the result of say_hello("Alice"), which is None because print functions return None.




```python
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("Something is happening before the function is called.")
        result = func(*args, **kwargs)
        print("Something is happening after the function is called.")
        return result
    return wrapper

@my_decorator
def say_hello(name):
    print(f"Hello, {name}!")

say_hello("Alice")
```

#Class decorators in Python are similar to function decorators, but they are used to modify or extend the behavior of classes. A class decorator is a function that takes a class as an argument and returns a new class or modifies the existing class.

```python
def class_decorator(cls):
    # Modify or extend the behavior of the class
    # Return the modified class
    pass
    return cls

@class_decorator
class MyClass:
    # Class definition
    pass
```

```python
def add_attributes(cls):
    cls.new_attribute = "I am a new attribute"
    cls.another_attribute = "Another new attribute"
    return cls
@add_attributes
class MyClass:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, my name is {self.name}"
obj = MyClass("Alice")
print(obj.greet())  # Output: Hello, my name is Alice
print(obj.new_attribute)  # Output: I am a new attribute
print(obj.another_attribute)  # Output: Another new attribute
```

#The @staticmethod decorator in Python is used to define a static method within a class. A static method is a method that operates independently of any instance of the class and does not require an instance to be passed as the first argument (self or cls). Static methods are typically used for utility functions or methods that perform operations unrelated to the state of any particular instance

```python
class MyClass:
    @staticmethod
    def static_method():
        # Method implementation
        pass
```

Static methods are useful when you want to define functions within a class that don't rely on instance-specific data or behavior. They are commonly used for utility functions, mathematical operations, data validation, or any functionality that doesn't require access to the instance's state.

```python
class MathUtils:
    @staticmethod
    def add(a, b):
        return a + b

    @staticmethod
    def subtract(a, b):
        return a - b

result = MathUtils.add(5, 3)
print(result)  # Output: 8

result = MathUtils.subtract(10, 4)
print(result)  # Output: 6
```
