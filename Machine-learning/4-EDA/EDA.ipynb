**What is Exploratory Data Analysis?**

Exploratory Data Analysis (EDA) is a method of analyzing datasets to understand their main characteristics. It involves summarizing data features, detecting patterns, and uncovering relationships through visual and statistical techniques. EDA helps in gaining insights and formulating hypotheses for further analysis.

#Step 3: Data Reduction
Some columns or variables can be dropped if they do not add value to our analysis.

In our dataset, the column S.No have only ID values, assuming they donâ€™t have any predictive power to predict the dependent variable.

axis=0: This indicates that the operation should be performed on rows. It is the default value.

axis=1: This indicates that the operation should be performed on columns.


```python
# Remove S.No. column from data
data = data.drop(['S.No.'], axis = 1)
data.info()
```

#Step 4: Univariate Analysis
Examine each variable individually.

```python
import matplotlib.pyplot as plt

# Distribution of 'Year'
data['Year'].hist()
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Year Distribution')
plt.show()

# Distribution of 'Kilometers_Driven'
data['Kilometers_Driven'].hist()
plt.xlabel('Kilometers Driven')
plt.ylabel('Frequency')
plt.title('Kilometers Driven Distribution')
plt.show()

# Distribution of 'Fuel_Type'
data['Fuel_Type'].value_counts().plot(kind='bar')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.title('Fuel Type Distribution')
plt.show()

# Distribution of 'Transmission'
data['Transmission'].value_counts().plot(kind='bar')
plt.xlabel('Transmission')
plt.ylabel('Count')
plt.title('Transmission Distribution')
plt.show()

# Distribution of 'Price'
data['Price'].hist()
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution')
plt.show()
```

#Step 5: Bivariate Analysis
Examine the relationships between pairs of variables.

```python
# Scatter plot of 'Year' vs 'Price'
data.plot.scatter(x='Year', y='Price')
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Year vs. Price')
plt.show()

# Scatter plot of 'Kilometers_Driven' vs 'Price'
data.plot.scatter(x='Kilometers_Driven', y='Price')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price')
plt.title('Kilometers Driven vs. Price')
plt.show()

# Box plot of 'Price' by 'Fuel_Type'
data.boxplot(column='Price', by='Fuel_Type')
plt.xlabel('Fuel Type')
plt.ylabel('Price')
plt.title('Price by Fuel Type')
plt.suptitle('')
plt.show()
```

#Step 6: Multivariate Analysis
Examine interactions among three or more variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select relevant columns for analysis
selected_columns = ['Price', 'Year', 'Kilometers_Driven', 'Fuel_Type']

# Subset the data
subset_data = data[selected_columns]

# Pairplot for multivariate analysis
sns.pairplot(subset_data, hue='Fuel_Type')
plt.suptitle('Pairplot for Multivariate Analysis', y=1.02)
plt.show()
```

Pairplot: This visualization shows pairwise relationships among the selected numerical variables ('Price', 'Year', 'Kilometers_Driven') with 'Fuel_Type' represented by different colors. It allows us to examine how these variables interact with each other across different fuel types.

#Step 7: Feature Engineering
Create new features if necessary.

```python
# Example: Create a new column for car age
current_year = 2024
data['Car_Age'] = current_year - data['Year']

# Display the updated DataFrame
print(data.head())
```

#Step 8: Reporting
Summarize findings and visualize key insights.

**Grouping and Aggregation:** We use groupby('Car_Age') to group the data by the 'Car_Age' column and then calculate the mean price (['Price'].mean()) for each group.

**Bar Plot:** We plot the average price against 'Car Age' using .plot(kind='bar'). This type of visualization helps in understanding how the average price varies with the age of cars.

```python
# Summary of findings
summary = data.describe()
print(summary)

# Visualization example: Average Price by Car Age
avg_price_by_age = data.groupby('Car_Age')['Price'].mean()
print(avg_price_by_age)
avg_price_by_age.plot(kind='bar')
plt.xlabel('Car Age')
plt.ylabel('Average Price')
plt.title('Average Price by Car Age')
plt.show()
```

#Exploratory Data Analysis in Python
Exploratory data analysis (EDA) is a critical initial step in the data science workflow. It involves using Python libraries to inspect, summarize, and visualize data to uncover trends, patterns, and relationships. Hereâ€™s a breakdown of the key steps in performing EDA with Python:

1. Importing Libraries:

pandas (pd): For data manipulation and analysis.
NumPy (np): For numerical computations.
Matplotlib.pyplot (plt): For basic plotting functionalities.
Seaborn (sns): A built-on top of Matplotlib, providing high-level visualization.
2. Loading the Data:

Use pd.read_csv() for CSV files, similar functions exist for other data formats (e.g., .xlsx, .json).
3. Initial Inspection:

Get an overview of the data using df.head(), .tail(), and .info().
Check data types with df.dtypes.
4. Data Cleaning:

Identify and handle missing values using methods like df.isnull().sum().
Find and address duplicates with df.duplicated().sum().
5. Univariate Analysis:

Analyze single variables at a time.
Use descriptive statistics with df.describe() for numerical data.
Create histograms, box plots, and density plots to visualize distributions.
6. Bivariate Analysis:

Explore relationships between two variables.
Create scatter plots to identify trends and potential correlations.
7. Visualization:

Effective visualizations are crucial for understanding data.
Use various plots like bar charts, pie charts, and heatmaps to represent categorical data.
