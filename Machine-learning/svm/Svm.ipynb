# Support Vector Machine (SVM) Algorithm
SVM is a powerful machine learning algorithm used for linear or nonlinear classification, regression, and even outlier detection tasks. SVMs can be used for a variety of tasks, such as text classification, image classification, spam detection, handwriting identification, gene expression analysis, face detection, and anomaly detection. SVMs are adaptable and efficient in a variety of applications because they can manage high-dimensional data and nonlinear relationships.

#hyperplane

The hyperplane tries that the margin between the closest points of different classes should be as maximum as possible.

 If the number of input features is two, then the hyperplane is just a line. If the number of input features is three, then the hyperplane becomes a 2-D plane

#Support Vector Machine Terminology
**Hyperplane**: Hyperplane is the decision boundary that is used to separate the data points of different classes in a feature space. In the case of linear classifications, it will be a linear equation i.e. wx+b = 0.

**Support Vectors**: Support vectors are the closest data points to the hyperplane, which makes a critical role in deciding the hyperplane and margin.

**Margin**: Margin is the distance between the support vector and hyperplane. The main objective of the support vector machine algorithm is to maximize the margin.  The wider margin indicates better classification performance.

#SVM implementation in Python
Predict if cancer is Benign or malignant. Using historical data about patients diagnosed with cancer enables doctors to differentiate malignant cases and benign ones are given independent attributes.

**Steps**
Load the breast cancer dataset from sklearn.datasets

Separate input features and target variables.

Buil and train the SVM classifiers using RBF kernel.

Plot the scatter plot of the input features.

Plot the decision boundary.



```python
# load_breast_cancer: Function to load the breast cancer dataset from Scikit-Learn's dataset library.
# matplotlib.pyplot as plt: Matplotlib library for plotting data.
# DecisionBoundaryDisplay: Utility for visualizing decision boundaries of classifiers.
# SVC: The Support Vector Classifier from Scikit-Learn.
```

```python
# 1.Load the important packages
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.svm import SVC
```

```python
# load_breast_cancer(): Loads the breast cancer dataset, which is a commonly used dataset for binary classification problems.
# X = cancer.data[:, :2]: Selects only the first two features (for easy visualization) from the dataset to use as the input features.
# y = cancer.target: Target labels for the dataset.
```

```python
# 2. Load the datasets
cancer = load_breast_cancer()
X = cancer.data[:, :2]
y = cancer.target
```

```python
# SVC: Initializes a Support Vector Classifier.
# kernel="rbf": Specifies that the Radial Basis Function (RBF) kernel should be used, which is effective for non-linear classification.
# gamma=0.5: Kernel coefficient for the RBF kernel. Controls the width of the Gaussian kernel.
# C=1.0: Regularization parameter. A higher value of C aims to classify all training examples correctly by giving the model more flexibility.
```

```python
#3 .Build the model
svm = SVC(kernel="rbf", gamma=0.5, C=1.0)
```

```python
#4. Trained the model
svm.fit(X, y)
```

```python
# DecisionBoundaryDisplay.from_estimator: Plots the decision boundary of the classifier.
# svm: The trained SVM model.
# X: The input features.
# response_method="predict": Uses the predict method of the SVM to determine the decision boundary.
# cmap=plt.cm.Spectral: Color map for the plot.
# alpha=0.8: Transparency of the decision boundary.
# xlabel and ylabel: Labels for the x and y axes, taken from the feature names of the breast cancer dataset.
```

```python
# 5 Plotting the Decision Boundary
DecisionBoundaryDisplay.from_estimator(
    svm,
    X,
    response_method="predict",
    cmap=plt.cm.Spectral,
    alpha=0.8,
    xlabel=cancer.feature_names[0],
    ylabel=cancer.feature_names[1],
)
# plt.scatter: Plots the data points on the graph.
# X[:, 0] and X[:, 1]: The first and second features of the dataset.
# c=y: Colors the points according to their class labels.
# s=20: Size of the points.
# edgecolors="k": Black edge color for the points.
# plt.show(): Displays the plot.
#6. Scatter Plot of the Data
plt.scatter(X[:, 0], X[:, 1], c=y, s=20, edgecolors="k")
plt.show()
```
