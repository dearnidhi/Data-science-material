```python
plt.subplot(1,2,1)
plt.plot(x,y,'g--')

plt.subplot(1,2,2)
plt.plot(y,x,'y*-')
```

```python
plt.subplot(1,4,1)
plt.plot(x,y,'g--')

plt.subplot(1,4,2)
plt.plot(x,y,'y*-')

plt.subplot(1,4,3)
plt.plot(x,y,'g--')

plt.subplot(1,4,4)
plt.plot(x,y,'y*-')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'g--')

plt.subplot(2,2,2)
plt.plot(x,y,'y*-')

plt.subplot(2,2,3)
plt.plot(x,y,'g--')

plt.subplot(2,2,4)
plt.plot(x,y,'y*-')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.5,0.7,1])
#add_axes(left,bottom,width,height)
axes.plot(x,y)
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.2,0.9,0.9])
#add_axes(left,bottom,width,height)
axes1.plot(x,y,'r',label="red")

axes2=fig.add_axes([0.15,0.75,0.3,0.3])
#add_axes(left,bottom,width,height)
axes2.plot(x,y,"b",label="blue")

axes1.legend(loc=4)
axes2.legend()
```

```python
x=[2,4,8]
y=[1,3,7]
x2=[11,13,15]
y2=[12,14,16]
plt.bar(x,y)
plt.bar(x2,y2)
plt.show()
```

```python
x=[2,4,8]
y=[1,3,7]
x2=[11,13,15]
y2=[12,14,16]
plt.bar(x,y)
plt.bar(x2,y2)
plt.title("bar graph")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
x=[2,4,8]
y=[1,3,7]
x2=[11,13,15]
y2=[12,14,16]
plt.barh(x,y)
plt.barh(x2,y2)
plt.title("bar graph")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```
